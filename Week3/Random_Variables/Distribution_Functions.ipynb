### Probability Mass Function (PMF)


```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats
from scipy.stats import norm
```

PMF is specific for discrete random variable (e.g., rolling a die). The following creates a PMF plot for the outcome of a six-sided die:

```python
data = np.arange(1,7)           # creates array of numbers from 1 to 6 that represents possible outcomes of rolling six-sided die
pmf = [1/6 for _ in data]       # creates list of probabilities for each outcome, due to 6 sided die has equal chance of landing on any of its faces, each outcome
                                # has a prob of 1/6. List comprehension that creates new list [] by performing operation 1/6 for each item in data list                    
plt.figure(figsize=(7,4))
plt.stem(data, pmf)             # creates stem plot often used to display PMFs with data array as x-values and pmf list as y-values
plt.title('PMF - Roll a die')
plt.xlabel('Die value')         
plt.ylabel('Probability')
plt.xticks(data)                # x-axis ticks corresponding to die values
```

### Probability Density Function (PDF)

This is specific for continuous random variable (e.g., height of people). Below we create a PDF plot for a standard normal distribution:

```python
x = np.linspace(-5,5,100)       # array of 100 evenly spaced numbers between -5 and 5 to represent x-values for the plot        
pdf = norm.pdf(x)               # calculates the PDF of a standard normal distribution at each of the x-values using norm.pdf function from scipy.stats
                                #...the PDF of a standard normal distribution is a bell-shaped curve that is symmetric around 0 
plt.figure(figsize=(7,4))
plt.plot(x, pdf, "r-")          # x array as x-values and pdf array as y-values, the "r-" arg specifies line as red and solid
plt.title('PDF - Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
```
